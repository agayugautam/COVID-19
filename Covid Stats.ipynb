```python
def cmr(deaths,population):
    cmr=(deaths/population)*100000
    print("Crude Mortality Rate ={}%".format(cmr))
```

```python
def ifr(deaths,confirmed):
    if(confirmed>0):
        ifr=(deaths/confirmed)*100
        print("IFR is ={}%".format(ifr))
```
